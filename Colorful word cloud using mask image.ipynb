```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
skill=open('skill.txt', 'r').read()
print("reading complete")
```

```python
stopwords=set(STOPWORDS)
```

```python
skillwc=WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

skillwc.generate(skill)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
fig = plt.figure()
fig.set_figwidth(10) # set width
fig.set_figheight(12) # set height

plt.imshow(skillwc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# save mask to mask_image
mask_image = np.array(Image.open('images/nepalmap.jpg'))
    
print('Image downloaded and saved!')
```

<a href="https://publicdomainpictures.net/en/">Link to download collections of mask image</a>

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(mask_image, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
skillwc = WordCloud(background_color='white', max_words=2000, mask=mask_image, stopwords=stopwords)

# generate the word cloud
skillwc.generate(skill)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(10) # set height

#these two lines of code are used to add the original color of image to word cloud
image_colors = ImageColorGenerator(mask_image)
skillwc.recolor(color_func=image_colors)

plt.imshow(skillwc, interpolation='bilinear')
plt.axis('off')
plt.show()
skillwc.to_file("Skill_word_cloud.png") #to save image

print("Loading complete")
```
